6. Explain the concept of generators in Python and how they are defined.
- In Python, a generator is a special type of function that creates an iterator, allowing you to generate a sequence of values on demand, one at a time, instead of storing the entire sequence in memory at once; it is defined by using the yield keyword within the function instead of return to produce each value in the sequence during iteration

```python
# Example
def fibonacci(n):

    a, b = 0, 1

    for _ in range(n):

        yield a

        a, b = b, a + b



# Usage

for num in fibonacci(10):

    print(num)
```

7. What are the advantages of using generators over regular functions?
- The primary advantage of using generators over regular functions is their ability to generate values on-demand, one at a time, which makes them significantly more memory efficient when dealing with large datasets or infinite sequences, as they don't need to store the entire sequence in memory at once; this is called "lazy evaluation" and can significantly improve performance, especially with large data sets.

```python
# Example
# generator function
def Square():

	number = 2

	#Create infinite loop
	while True:
		# Yield the current value of 'number'
		yield number

		# Calculate the square of 'number' and update its value
		number *= number

# Create a generator object 'Sq' by calling the 'Square()' generator function
Sq = Square()

# Function call
print(next(Sq))


print(next(Sq))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- In Python, "map()" applies a function to each element of an iterable, creating a new iterable with the transformed elements;
- "filter()" selects elements from an iterable based on a condition, returning only those that satisfy the condition; and
- "reduce()" combines all elements of an iterable into a single value using a specified function, accumulating the result iteratively.

```python
# Example
from functools import reduce
numbers = [1, 2, 3, 4, 5]


# map: square each number

squared_numbers = map(lambda x: x**2, numbers)

print(list(squared_numbers))



# filter: select even numbers

even_numbers = filter(lambda x: x % 2 == 0, numbers)

print(list(even_numbers))



# reduce: calculate the sum of all numbers

total_sum = reduce(lambda x, y: x + y, numbers)

print(total_sum)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Checking prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.x, self.y = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        self.count += 1
        fib_number = self.x
        self.x, self.y = self.y, self.x + self.y
        return fib_number

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage
for power in powers_of_2(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.


def read_file(file_path):
    with open(file_path, 'r') as file:
        yield from file  # Yields each line directly

# Example usage
for line in read_file("example.txt"):
    print(line.strip())  # Remove extra spaces or newlines
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Convert to Fahrenheit using map and lambda
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Print the converted temperatures
print(fahrenheit_temps)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
    return "".join(filter(lambda x: x.lower() not in "aeiou", string))

# Example usage
text = "Hello, how are you?"
result = remove_vowels(text)
print(result)
```
